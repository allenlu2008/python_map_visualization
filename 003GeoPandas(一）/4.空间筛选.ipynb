```python
import geopandas,pandas
```

```python
import shapely.geometry as geom
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.sans-serif']=['Songti SC']
mpl.rcParams['axes.unicode_minus'] = False 
```

```python
bj = geopandas.read_file("./data/beijing/行政区划.shp")
```

```python
bj
```

```python
poly = geom.Polygon([(116.31,39.85),(116.45,39.85),(116.45,40),(116.31,40)])
```

```python
poly
```

```python
it = bj[bj.geometry.intersects(poly)]
it
```

```python
fig,ax = plt.subplots(figsize=(10,10))
it.plot(ax=ax,cmap="tab10")
geopandas.GeoDataFrame(pandas.DataFrame({"geometry":[poly]}).set_geometry("geometry")).plot(ax=ax,
                                                                                facecolor="white",hatch="\\",
                                                                                           alpha=0.2)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
bj[bj.geometry.withinin(poly)].plot(ax=ax,cmap="tab10")
geopandas.GeoDataFrame(pandas.DataFrame({"geometry":[poly]}).set_geometry("geometry")).plot(ax=ax,
                                                                                facecolor="white",hatch="\\",
                                                                                           alpha=0.2)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
bj[bj.geometry.overlaps(poly)].plot(ax=ax,cmap="tab10")
geopandas.GeoDataFrame(pandas.DataFrame({"geometry":[poly]}).set_geometry("geometry")).plot(ax=ax,
                                                                                facecolor="white",hatch="\\",
                                                                                           alpha=0.2)
```
