```python
import matplotlib.pyplot as plt
```

```python
mk = [".",",","o","v","^",">","<","1","2","3","4","8","s","p","*",
     "h","H","+","x","D","d","|","_"]
```

```python
plt.figure(figsize=(9,7))
for i in range(1,len(mk)+1):
    plt.scatter(i,0,s=200, marker=mk[i-1])
    if i % 2 ==0:
        plt.text(i,-0.0025,s=mk[i-1],fontsize=20)
    else:
        plt.text(i,0.0025,s=mk[i-1],fontsize=20)
```
