```python
import matplotlib.pyplot as plt
import matplotlib.offsetbox as os
```

```python
import random
import PIL.Image as Image
```

```python
x = [random.randint(1,100) for i in range(50)]
y = [random.randint(1,100) for i in range(50)]
```

```python
plt.figure(figsize=(9,6))
plt.scatter(x,y)
```

```python
from matplotlib._png import read_png
```

```python
fig,ax = plt.subplots(figsize=(9,6))
ax.scatter(x,y)
for x1,y1 in zip(x,y):
    image = os.OffsetImage(read_png("image/{0}.png".format(random.randint(1,6))))
    img = os.AnnotationBbox(image,(x1,y1),frameon=False)
    ax.add_artist(img)
```

```python
import numpy
```

```python
x = numpy.linspace(0,2*numpy.pi,100)
y = numpy.sin(x)
```

```python
fig,ax = plt.subplots(figsize=(12,6))
ax.plot(x,y,zorder=-1)
for x1,y1 in zip(x[::5],y[::5]):
    image = os.OffsetImage(read_png("image/{0}.png".format(random.randint(1,6))))
    img = os.AnnotationBbox(image,(x1,y1),frameon=False)
    ax.add_artist(img)
```
