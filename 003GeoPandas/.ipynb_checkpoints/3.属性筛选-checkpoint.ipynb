```python
import geopandas
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.sans-serif']=['Songti SC']
mpl.rcParams['axes.unicode_minus'] = False 
```

```python
bj = geopandas.read_file("./data/beijing/行政区划.shp")
```

```python
bj
```

```python
qu = bj[bj["NAME"].str.contains("区")]
```

```python
fig,ax = plt.subplots(figsize=(10,10))
qu.plot(ax=ax,cmap="tab20")
```

```python
clq = bj[bj["面积"] < 500]
```

```python
clq
```

```python
fig,ax = plt.subplots(figsize=(10,10))
clq.plot(ax=ax,cmap="tab20")
```

```python
fig,ax = plt.subplots(figsize=(10,10))
clq.plot(ax=ax,cmap="tab10")
for idx,q in clq.iterrows():
    ax.text(q.geometry.centroid.x-0.02,q.geometry.centroid.y,s=q["NAME"],
           color="w",fontsize=12)
```

```python
dx = bj[(bj["NAME"] == "东城区") | (bj["NAME"]=="西城区")]
```

```python
dx.geometry.bounds
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.set_xlim(min(dx.geometry.bounds["minx"]),max(dx.geometry.bounds["maxx"]))
ax.set_ylim(min(dx.geometry.bounds["miny"]),max(dx.geometry.bounds["maxy"]))
bj.plot(ax=ax,facecolor="white",hatch="\\")
dx.plot(ax=ax,cmap="tab10")
```
