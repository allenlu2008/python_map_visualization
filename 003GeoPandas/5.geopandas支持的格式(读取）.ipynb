## postgis

```python
from sqlalchemy import create_engine
```

#### 创建数据库连接

```python
db_connection_url = "postgres://postgis:postgis@127.0.0.1:5432/postgis";
engine = create_engine(db_connection_url)
```

#### 通过SQL，来读取空间数据

```python
sql = "select * from world.world_chinese"
world = geopandas.read_postgis(sql=sql, con=engine)
```

```python
world.head()
```

```python
world.plot(figsize=(9,8),cmap="tab20")
```
